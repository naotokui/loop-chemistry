```python
from glob import glob
import shutil
import os
from tqdm import tqdm

target_folder =  '/media/nao/53ECDE8E77C8F287/Mosaic_video/YouTube-5M/videos'
files = glob("YouTube-music-video-5M/*/*")

for f in tqdm(files):
    artist = f.split("/")[1]
    basename = os.path.basename(f)
    target_path = os.path.join(target_folder, artist + "__" + basename)
    if os.path.exists(target_path) == False:
        shutil.copy(f, target_path)
    
```
